4- run the simulation for <strong>100 sec</strong><br>

Start by executing next code to import nengo and the different libraries needed for the task

```python
import nengo 
import numpy as np 
import matplotlib.pyplot as plt
```

Your Code:

```python
# initiate the network
myNetwork = nengo.Network("Nodes Network")
# add Node and Probe
with myNetwork:
    weirdIn = nengo.Node(lambda t : t * np.sin(t * np.pi) if(t<50) else (100-t) * np.sin((100-t) * np.pi))
    nodeProbe = nengo.Probe(weirdIn)

# run the simulation
with nengo.Simulator(myNetwork, progress_bar=False) as sim:
    sim.run(100)
```

Show the output of your Node

```python
# use matplotlib
plt.plot(sim.data[nodeProbe])
```
